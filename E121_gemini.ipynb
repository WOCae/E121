```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
```

```python
# 文字を数値に変換する辞書
char_to_int = {'み': 0, 'か': 1, 'た': 2}
int_to_char = dict((i, c) for i, c in enumerate(char_to_int))

# 入力データと出力データの作成
data = ['みか', 'かた', 'みみ']
X = []
y = []
for word in data:
    for i in range(len(word)-1):
        X.append(char_to_int[word[i]])
        y.append(char_to_int[word[i+1]])

# numpy配列に変換
X = np.reshape(X, (len(X), 1))
y = np.eye(len(char_to_int))[y]
```

```python
# モデルの構築
model = Sequential()
model.add(SimpleRNN(units=3, input_shape=(1, 1)))
model.add(Dense(len(char_to_int), activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
model.fit(X, y, epochs=100, verbose=2)
```

```python
# 予測する文字列
text = "た"
for _ in range(2):
    x = np.reshape(char_to_int[text[-1]], (1, 1))
    prediction = model.predict(x)
    index = np.argmax(prediction)
    result = int_to_char[index]
    text += result
print(text)
```
